```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Datensatz einlesen
data = pd.read_csv('schlaf_gesundheit_lifestyle_datensatz.csv')

# 1. Überblick über die Daten
print(data.info())
print(data.head())

# 2. Streudiagramm: Alter vs. Schlafdauer
plt.figure()
plt.scatter(data['Alter'], data['Schlafdauer'])
plt.xlabel('Alter [Jahre]')
plt.ylabel('Schlafdauer [Stunden]')
plt.title('Zusammenhang zwischen Alter und Schlafdauer')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Liniendiagramm: Durchschnittliche Herzfrequenz pro Altersgruppe
herzfrequenz_pro_alter = data.groupby('Alter')['Herzfrequenz'].mean()

plt.figure()
plt.plot(herzfrequenz_pro_alter.index, herzfrequenz_pro_alter.values, 
         color='red', marker='o', linewidth=2)
plt.xlabel('Alter [Jahre]')
plt.ylabel('Durchschnittliche Herzfrequenz [Schläge/min]')
plt.title('Herzfrequenz nach Altersgruppen')
plt.grid(True, alpha=0.3)
#plt.savefig('herzfrequenz_alter.png', dpi=300, bbox_inches='tight')
plt.show()
```
